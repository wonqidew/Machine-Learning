### Описание датасета:
Данный датасет представляет собой информацию о пропорции добавленной стоимости от высокотехнологичной и средней промышленности к общей добавленной стоимости за период с 1990 по 2015 год. Данные являются реальными и взяты с открытого источника Kaggle (https://www.kaggle.com/datasets).

### Описание атрибутов датасета:
В выбранном датасете представлены два атрибута:
- **Год**: представляет собой год, к которому относится пропорция добавленной стоимости. Тип данных - числовой.
- **Пропорция**: представляет собой значение пропорции добавленной стоимости от высокотехнологичной и средней промышленности к общей добавленной стоимости. Тип данных - числовой.

### Описание задачи анализа:
Задача анализа данного датасета заключается в изучении изменения пропорции добавленной стоимости от высокотехнологичной и средней промышленности к общей добавленной стоимости со временем.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Чтение данных из CSV-файла
df = pd.read_csv('hightech.csv', delimiter=';')

# Среднее значение и стандартное отклонение для атрибута "Год"
mean_year = df['year'].mean()
std_year = df['year'].std()
print("Среднее значение для атрибута 'year': ", mean_year)
print("Стандартное отклонение для атрибута 'year': ", std_year)

# Среднее значение и стандартное отклонение для атрибута "Пропорция"
mean_proportion = df['proportion'].mean()
std_proportion = df['proportion'].std()
print("Среднее значение для атрибута 'proportion': ", mean_proportion)
print("Стандартное отклонение для атрибута 'proportion': ", std_proportion)
```

```python
# Гистограмма распределения значений атрибута "Пропорция"
plt.hist(df['proportion'], bins=10)
plt.xlabel('Пропорция')
plt.ylabel('Частота')
plt.title('Гистограмма распределения пропорции')
plt.show()

# Гистограмма распределения значений атрибута "Год"
plt.hist(df['year'], bins=10)
plt.xlabel('Год')
plt.ylabel('Частота')
plt.title('Гистограмма распределения года')
plt.show()
```
